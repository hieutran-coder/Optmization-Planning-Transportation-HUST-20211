```python
import time
class TimeCounter():
  def __init__(self):
    self.start_time = 0
    self.end_time = 0
  def start(self):
    self.start_time = time.time()
  def end(self):
    self.end_time = time.time()
  def get_duration(self):
    return self.end_time - self.start_time
```

```python
solver = pywraplp.Solver.CreateSolver('SCIP')
```

```python
# định nghĩa biến
t = {}
for i in range(m):
    for j in range(n):
        t[i,j] = solver.IntVar(0, 1, 't[' + str(i) +',' + str(j)+ ']')


    
```

```python
# Tạo constraint cho các biến t[i][j]: Loại những hub ko thuộc 5 hub gần nhất với mỗi khách hàng
for i in range(m):
    for j in range(n):
        if( a[i][j] == 0):
            const1 = solver.Add(t[i,j] == 0)
print('Number of constraints =', solver.NumConstraints())


```

```python
# Tạo constraint cho tổng số đơn hàng của 1 khách hàng tại tất cả hub luôn là v_c
for i in range(m):
            solver.Add(solver.Sum(a[i][j]*v_c[i]*t[i,j] for j in range(n)) == v_c[i])
print('Number of constraints =', solver.NumConstraints())
```

```python
# Tạo constraint tổng số đơn hàng trên mỗi hub ko vượt quá hub đó. 
for j in range(n):
            solver.Add(solver.Sum(a[i][j]*v_c[i]*t[i,j] for i in range(m)) <= v_h[j])
print('Number of constraints =', solver.NumConstraints())
```

```python
objective_terms = []
for i in range(m):
    for j in range(n):
        objective_terms.append(a[i][j]*d[i][j]*t[i,j])
solver.Minimize(solver.Sum(objective_terms))
```

```python
COUNTER.start()
status = solver.Solve()
COUNTER.end()
```

```python
if status == pywraplp.Solver.OPTIMAL or pywraplp.Solver.FEASIBLE:
    print('Solution:')
    print('Objective value = ', solver.Objective().Value())
else: print('not found solution')
a=0
for i in range(m):
    for j in range(n):
        if (t[i,j].solution_value() == 1):
            a = a + d[i][j]
            print(i,j,d[i][j])
print(a)
objective_output= str(solver.Objective().Value())

h = [0]*m
for i in range(m):
    for j in range(n):
        if (t[i,j].solution_value() == 1):
                h[i] = j+1

print(h) 
print("Thời gian chạy thuật toán MIP là",COUNTER.get_duration())       
```
